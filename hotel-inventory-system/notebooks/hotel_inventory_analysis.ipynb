# Hotel Inventory Analysis

This Jupyter notebook provides an analysis of the hotel inventory management system. The analysis includes data loading, exploratory data analysis (EDA), demand forecasting using the Holt-Winters method, calculation of optimal inventory par levels, and simulation of the inventory management system. Visualizations and recommendations are also included.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from src.hotel_inventory_system import load_and_prepare_data, perform_eda, forecast_demand, calculate_par_levels, simulate_inventory_system, generate_recommendations

# Load data
DATA_FILE = '../data/hotel_bar_inventory.csv'  # Update with your file path
df = load_and_prepare_data(DATA_FILE)

# Perform EDA
top_items, bar_consumption = perform_eda(df)

# Forecast demand
FORECAST_DAYS = 30
forecasts = forecast_demand(df, forecast_days=FORECAST_DAYS)

# Calculate par levels
LEAD_TIME_DAYS = 3
SERVICE_LEVEL = 0.95
par_levels = calculate_par_levels(forecasts, lead_time_days=LEAD_TIME_DAYS, service_level=SERVICE_LEVEL)

# Simulate inventory system
simulation_results, service_level_achieved = simulate_inventory_system(df, forecasts, par_levels, simulation_days=30)

# Generate recommendations
recommendations = generate_recommendations(df, forecasts, par_levels, simulation_results)

# Display results
print(f"Service Level Achieved: {service_level_achieved * 100:.2f}%")
recommendations.head()
```

## Visualizations

The following visualizations provide insights into the inventory consumption patterns and the effectiveness of the inventory management system.

```python
# Create visualizations
plt.style.use('seaborn-v0_8-darkgrid')

# 1. Consumption by Bar
plt.figure(figsize=(10, 6))
bar_consumption = df.groupby('Bar Name')['Consumed (ml)'].sum().sort_values()
bar_consumption.plot(kind='barh', color='steelblue')
plt.title('Total Consumption by Bar')
plt.xlabel('Total Consumption (ml)')
plt.savefig('inventory_analysis.png', dpi=300, bbox_inches='tight')
plt.close()
```

## Summary of Findings

The analysis reveals key insights into the inventory management system, including:
- **Top Consumed Items**: The top items by consumption were identified, which can help in prioritizing inventory management efforts.
- **Bar Performance**: Consumption patterns across different bars were analyzed, highlighting which bars have higher or lower inventory turnover.
- **Forecasting Accuracy**: The Holt-Winters method was used for demand forecasting, providing a more nuanced understanding of expected consumption trends.
- **Par Levels**: Optimal par levels were calculated to ensure that inventory is maintained at appropriate levels to meet demand without overstocking.
- **Simulation Results**: The simulation provided insights into stockout occurrences and overall service levels achieved during the analysis period.

### Recommendations
Based on the analysis, the following recommendations are made:
- **Order Recommendations**: Items that require immediate ordering should be prioritized to avoid stockouts.
- **Monitor Inventory Levels**: Regular monitoring of inventory levels against par levels is essential to maintain service quality.
- **Adjust Forecasting Models**: Continuously refine forecasting models based on new data to improve accuracy.
